```python
import pandas as pd

X_train = pd.read_csv('FIFA_X_train.csv', encoding='cp949')
X_test = pd.read_csv('FIFA_X_test.csv', encoding='cp949')
y_train = pd.read_csv('FIFA_y_train.csv', encoding='cp949')
y_test = pd.read_csv('FIFA_y_test.csv', encoding='cp949')

# print(X_train.info())
# print(X_train.head())
# print(y_train.head())
# print(X_train.describe())

ID = X_test['ID'].copy()

# ID 컬럼 삭제
X_train = X_train.drop('ID', axis = 1)
X_test = X_test.drop('ID', axis = 1)
y_train = y_train.drop('ID', axis = 1)
y_test = y_test.drop('ID', axis = 1)


# Position_Class 변수 재생성
X_train_Position_Class = []
X_test_Position_Class = []
Forward = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW']
Midfielder = ['LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM']
Defender = ['LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
GoalKeeper = ['GK']

for i in X_train['Position']:
    if i in Forward:
        X_train_Position_Class.append('Forward')
    elif i in Midfielder:
        X_train_Position_Class.append('Midfielder')
    elif i in Defender:
        X_train_Position_Class.append('Defender')
    elif i in GoalKeeper:
        X_train_Position_Class.append('GoalKeeper')

for i in X_test['Position']:
    if i in Forward:
        X_test_Position_Class.append('Forward')
    elif i in Midfielder:
        X_test_Position_Class.append('Midfielder')
    elif i in Defender:
        X_test_Position_Class.append('Defender')
    elif i in GoalKeeper:
        X_test_Position_Class.append('GoalKeeper')

X_train['Position_Class'] = X_train_Position_Class
X_test['Position_Class'] = X_test_Position_Class

# print(X_train['Position_Class'].value_counts())

# Position 컬럼 삭제
X_train.drop('Position', axis=1, inplace=True)
X_test.drop('Position', axis=1, inplace=True)

# Jersey Number 컬럼 삭제
X_train.drop('Jersey_Number', axis=1, inplace=True)
X_test.drop('Jersey_Number', axis=1, inplace=True)


# Height_cm 변수 재생성
X_train_Height_cm = []
X_test_Height_cm = []
for i in X_train['Height']:
    fit, inch = i.split("'")
    X_train_Height_cm.append(int(fit)*30 + int(inch)*2.5)

for i in X_test['Height']:
    fit, inch = i.split("'")
    X_test_Height_cm.append(int(fit)*30 + int(inch)*2.5)

X_train['Height_cm'] = X_train_Height_cm
X_test['Height_cm'] = X_test_Height_cm


# Height 변수 제거
X_train.drop('Height', axis=1, inplace=True)
X_test.drop('Height', axis=1, inplace=True)

# print(X_train.info())
# print(X_test.info())

# Weight_lb 결측값 평균으로 대체
Weight_lb_mean = X_train['Weight_lb'].mean()
X_train['Weight_lb'].fillna(Weight_lb_mean, inplace=True)

# print(X_train.info())

# Club 컬럼 삭제
X_train.drop('Club', axis=1, inplace=True)
X_test.drop('Club', axis=1, inplace=True)

# print(X_train.info())
# print(X_train['Age'].unique())

# Age 변수 범주화
X_train['Age'] = X_train['Age'].apply(lambda x : int(x[0]))
X_test['Age'] = X_test['Age'].apply(lambda x : int(x[0]))

# print(X_train.info())
# print(X_train['Work_Rate'])

# 공격 운동량, 방어 운동량 변수 생성
X_train_AWR = []
X_test_AWR = []
X_train_DWR = []
X_test_DWR = []
for i in X_train['Work_Rate']:
    AWR, DWR = i.split('/ ')
    X_train_AWR.append(AWR)
    X_train_DWR.append(DWR)

for i in X_test['Work_Rate']:
    AWR, DWR = i.split('/ ')
    X_test_AWR.append(AWR)
    X_test_DWR.append(DWR)

X_train['AWR'] = X_train_AWR
X_train['DWR'] = X_train_DWR
X_test['AWR'] = X_test_AWR
X_test['DWR'] = X_test_DWR

# Work Rate 컬럼 제거
X_train.drop('Work_Rate', axis=1, inplace=True)
X_test.drop('Work_Rate', axis=1, inplace=True)

# print(X_train.info())
# print(y_train)
# print(X_train['Nationality'].value_counts()) # 범주가 너무 많다

# Nationality 컬럼 제거
X_train.drop('Nationality', axis=1, inplace=True)
X_test.drop('Nationality', axis=1, inplace=True)

# print(X_train.info())

# print(X_train['Preferred_Foot'].value_counts())
preferred_foot_dict = {'Right' : 0, 'Left' : 1}
X_train['Preferred_Foot'] = X_train['Preferred_Foot'].apply(lambda x : preferred_foot_dict[x])
X_test['Preferred_Foot'] = X_test['Preferred_Foot'].apply(lambda x : preferred_foot_dict[x])

# print(X_train.info())
# print(X_train['Position_Class'].value_counts())

X_train_position_class_OH = pd.get_dummies(X_train['Position_Class'])
X_test_position_class_OH = pd.get_dummies(X_test['Position_Class'])

# print(X_train)
# print(X_train_position_class_OH)

new_X_train = pd.concat([X_train.drop('Position_Class', axis=1), X_train_position_class_OH], axis=1)
new_X_test = pd.concat([X_test.drop('Position_Class', axis=1), X_test_position_class_OH], axis=1)
# print(new_X_train.info())
# print(new_X_test.info())

# print(new_X_train['AWR'].value_counts())
# print(new_X_train['DWR'].value_counts())

# 범주형 변수 원핫 인코딩
X_train_AWR_OH = pd.get_dummies(new_X_train['AWR'], prefix='AWR')
X_test_AWR_OH = pd.get_dummies(new_X_test['AWR'], prefix='AWR')
X_train_DWR_OH = pd.get_dummies(new_X_train['DWR'], prefix='DWR')
X_test_DWR_OH = pd.get_dummies(new_X_test['DWR'], prefix='DWR')

new_X_train = pd.concat([new_X_train, X_train_AWR_OH, X_train_DWR_OH], axis=1)
new_X_test = pd.concat([new_X_test, X_test_AWR_OH, X_test_DWR_OH], axis=1)

new_X_train.drop('AWR', axis=1, inplace=True)
new_X_train.drop('DWR', axis=1, inplace=True)
new_X_test.drop('AWR', axis=1, inplace=True)
new_X_test.drop('DWR', axis=1, inplace=True)

# print(new_X_train.info())

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X_trainDF, X_validDF, y_trainSR, y_validSR = train_test_split(new_X_train, y_train, random_state=42)
# print(X_trainDF.shape, X_validDF.shape)

y_trainSR = y_trainSR['Value']
y_validSR = y_validSR['Value']

rf = RandomForestRegressor(random_state=42)

rf.fit(X_trainDF, y_trainSR)

pred_y = rf.predict(X_validDF)

from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
r2 = r2_score(y_validSR, pred_y)
mse = mean_squared_error(y_validSR, pred_y)
rmse = np.sqrt(mse)
print(f"r2 : {r2}")
print(f"rmse : {rmse}")

pred_y_test = rf.predict(new_X_test)

print(r2_score(y_test, pred_y_test))
mse_test = mean_squared_error(y_test, pred_y_test)
rmse_test = np.sqrt(mse_test)
print(rmse_test)

result = pd.DataFrame({'ID' : ID,
                       'Value' : pred_y_test})
# print(result)

result.to_csv('12345.csv', index=False)
```
